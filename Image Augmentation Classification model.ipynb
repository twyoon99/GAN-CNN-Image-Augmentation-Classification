#1. 원본 데이터만 갖고 돌리기

##전처리

```python
data_dir = ['/content/og/bacteria','/content/og/normal','/content/og/virus'] # 파일 경로 설정
input_size = (128, 128) # 이미지를 Array로 읽어옴
```

```python
bacteria_shuffle = random.sample(range(2780),1493)
normal_shuffle = random.sample(range(1575),1493)
virus_shuffle = random.sample(range(1493),1493)
```

```python
bac = os.listdir(data_dir[0])
bac = np.array(bac)[bacteria_shuffle].tolist()

nor = os.listdir(data_dir[1])
nor = np.array(nor)[normal_shuffle].tolist()

vir = os.listdir(data_dir[2])

shuffle_dir=[[data_dir[0],bac],[data_dir[1],nor],[data_dir[2],vir]]
```

```python
total_y = np.array([0]*1493 + [1]*1493 + [2]*1493)
```

```python
#전처리
cnt = 0
total_x = np.zeros((4479, 128, 128, 3), dtype=np.float32)
for j in shuffle_dir:
    for i, img_name in enumerate(j[1]):
        img_path = os.path.join(j[0], img_name)
        print(img_path)
        img = Image.open(img_path).convert('RGB')
        print(type(img))

        # Image resize
        img = img.resize(input_size)

        # Image Normalize
        img = np.array(img, dtype=np.float32) / 255.0

        # Add the image to the arrays
        total_x[cnt] = img

        cnt += 1
        print(cnt)
```

```python
x_train, x_test, y_train, y_test = train_test_split(total_x, total_y, test_size=0.15)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2)
```

##Model, Train, Test

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(5, 5), strides=(1, 1), padding='same', activation='relu', input_shape=[128, 128, 3]))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(64, (2, 2), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))
print(model.summary())

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
earlyStopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, verbose=0, mode='min')
hist = model.fit(x_train, y_train, batch_size=128, epochs=100, verbose=1, validation_data=(x_val, y_val), callbacks=[earlyStopping])
```

#2. 생성 데이터도 합쳐서 돌리기

##전처리

```python
data_dir = ['/content/og/bacteria','/content/content/bacteria','/content/og/normal','/content/content/normal','/content/og/virus','/content/content/virus']
input_size = (128, 128)
```

```python
bac = os.listdir(data_dir[0])
new_bac = os.listdir(data_dir[1])

nor = os.listdir(data_dir[2])
new_nor = os.listdir(data_dir[3])

vir = os.listdir(data_dir[4])
new_vir = os.listdir(data_dir[5])

shuffle_dir=[[data_dir[0],bac],[data_dir[1],new_bac],[data_dir[2],nor],[data_dir[3],new_nor],[data_dir[4],vir],[data_dir[5],new_vir]]
```

```python
new_total_y = np.array([0]*2999 + [1]*3000 + [2]*2999)
```

```python
#전처리
cnt = 0
new_total_x = np.zeros((8998, 128, 128, 3), dtype=np.float32)
for j in shuffle_dir:
    for i, img_name in enumerate(j[1]):
        img_path = os.path.join(j[0], img_name)
        print(img_path)
        img = Image.open(img_path).convert('RGB')
        print(type(img))

        # Image resize
        img = img.resize(input_size)

        # Image Normalize
        img = np.array(img, dtype=np.float32) / 255.0

        # Add the image to the arrays
        new_total_x[cnt] = img

        cnt += 1
        print(cnt)
```

```python
new_x_train, new_x_test, new_y_train, new_y_test = train_test_split(new_total_x, new_total_y, test_size=0.15)
new_x_train, new_x_val, new_y_train, new_y_val = train_test_split(new_x_train, new_y_train, test_size=0.2)
```

##Model, Train, Test

```python
new_model = Sequential()
new_model.add(Conv2D(32, kernel_size=(5, 5), strides=(1, 1), padding='same', activation='relu', input_shape=[128, 128, 3]))
new_model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
new_model.add(Conv2D(64, (2, 2), activation='relu', padding='same'))
new_model.add(MaxPooling2D(pool_size=(2, 2)))
new_model.add(Dropout(0.25))
new_model.add(Flatten())
new_model.add(Dense(1000, activation='relu'))
new_model.add(Dropout(0.5))
new_model.add(Dense(3, activation='softmax'))
print(new_model.summary())

new_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
new_earlyStopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, verbose=0, mode='min')
new_hist = new_model.fit(new_x_train, new_y_train, batch_size=128, epochs=100, verbose=1, validation_data=(new_x_val, new_y_val), callbacks=[earlyStopping])
```

#3. Evaluate
